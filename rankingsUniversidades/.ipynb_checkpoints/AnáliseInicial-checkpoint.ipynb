```python
rTimes2012 = timesData[timesData['year'] == 2012]
rTimes2012
```

```python
rTimes2012.index = list(range(402))
rTimes2012[45:48]
```

```python
rTimes2012 = timesData[timesData['year'] == 2012]
rTimes2012.index = range(len(rTimes2012))
for i in range(len(rTimes2012)):
    if (not pd.isna(rTimes2012.iloc[i,11])):
        valorNovo = rTimes2012.at[i,'international_students']
        if '%' in valorNovo:
            valorNovo = valorNovo[:-1]
        if valorNovo.strip() not in ['']:
            rTimes2012.at[i,'international_students'] = int(valorNovo)
rTimes2012[0:200].plot.scatter(x=0,y=11,xticks=range(0,200,25))
```
